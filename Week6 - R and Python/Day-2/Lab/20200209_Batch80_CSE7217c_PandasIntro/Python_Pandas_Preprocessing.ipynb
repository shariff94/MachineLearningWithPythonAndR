## Selection (also called indexing)

### Selection by Position

.iloc is primarily integer position based (from 0 to length of the axis -1 ), but may also be used with a boolean array.

.iloc will raise IndexError if a requested indexer is out-of-bounds, except slice indexers which allow out-of-bounds indexing.

```python
df.iloc[3:5,0:2]
```

### Selection by Label

- .loc is primarily label based, but may also be used with a boolean array. .loc will raise KeyError when the items are not found. 

- label slicing, both endpoints are included

```python
df.loc[range(3),['A','B']]
```

# Joins

```python
data1 = pd.read_csv("Customer_sample_1.csv")
data2 = pd.read_csv("Customer_sample_2.csv")
```

```python
data1
```

```python
data2
```

## Inner Join

```python
data_ij = pd.merge(left=data1, right=data2, how='inner', 
                   left_on='ID', right_on='Customer ID')
```

```python
data_ij
```

```python
data_ij.shape
```

## Left Join

```python
data_lj = pd.merge(left=data1, right=data2, how='left', 
                   left_on='ID', right_on='Customer ID')
```

```python
data_lj
```

## Right Join

```python
data_rj = pd.merge(left=data1, right=data2, how='right', left_on='ID', right_on='Customer ID')
```

```python
data_rj
```

## Full Join / Full Outer Join

```python
data_fj = pd.merge(left=data1, right=data2, how='outer', left_on='ID', right_on='Customer ID')
```

```python
data_fj
```

## Performing Data TypeCasting

```python
data_fj['ZIP Code'] = data_fj['ZIP Code'].astype('category')
```

```python
data_fj.dtypes
```

```python
# convert the required variables into category in a single shot
cat_cols = ['ID', 'Personal Loan', 'Customer ID', 'ZIP Code']

for i in cat_cols:
    data_fj[i] = data_fj[i].astype('category')
```

```python
data_fj.dtypes
```
